```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12,5)
```

```python
df = pd.read_csv('data/superstore_sales.csv', encoding="latin1")
df['Order Date'] = pd.to_datetime(df['Order Date'], infer_datetime_format=True, errors='coerce')
df.head()
```

```python
df.head()
```

```python
#Check structure
df.info()

#Summary statistics
df.describe()

#Check missing values
df.isnull().sum()
```

```python
#Clean data

#Convert Order Date column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors="coerce")

#Drop rows with invalid/missing dates
df = df.dropna(subset=["Order Date"])

#Verify
df['Order Date'].head()




```

```python
#Monthly Sales Trend

monthly_sales=(
    df.groupby(df['Order Date'].dt.to_period('M'))['Sales']
    .sum()
    .reset_index()
)
monthly_sales['Order Date'] = monthly_sales['Order Date'].dt.to_timestamp()

#Plot
plt.plot(monthly_sales['Order Date'], monthly_sales['Sales'], marker='o')
plt.title("Monthly Sales Trend (2014-2017)")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.savefig("monthly_sales.png", dpi=300, bbox_inches="tight") 
plt.show()
```

```python
# Sales by Category
category_sales = df.groupby('Category')['Sales'].sum().reset_index()

sns.barplot(x='Category', y='Sales', data=category_sales, hue='Category', palette="viridis", legend=False)
plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.savefig("category_sales.png", dpi=300, bbox_inches="tight") 
plt.show()
```

```python
# Sales by Region

region_sales = df.groupby('Region')['Sales'].sum().reset_index()
sns.barplot(x='Region', y='Sales', data=region_sales, hue='Region', palette='muted', legend=False)
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.savefig("region_sales.png", dpi=300, bbox_inches="tight") 
plt.show()
```

```python
# Profit vs. Sales by Month

monthly_perf = (
    df.groupby(df['Order Date'].dt.to_period('M'))[['Sales', 'Profit']]
    .sum()
    .reset_index()
)
monthly_perf['Order Date'] = monthly_perf['Order Date'].dt.to_timestamp()

# Dual-axis chart
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(monthly_perf['Order Date'], monthly_perf['Sales'], color="skyblue", alpha=0.6, label="Sales")
ax2.plot(monthly_perf['Order Date'], monthly_perf['Profit'], color="red", marker="o", label="Profit")

ax1.set_xlabel("Month")
ax1.set_ylabel("Sales", color="blue")
ax2.set_ylabel("Profit", color="red")
plt.title("Profit vs. Sales by Month")
fig.autofmt_xdate()
plt.savefig("profit_vs_sales_by_month.png", dpi=300, bbox_inches="tight") 
plt.show()
```
